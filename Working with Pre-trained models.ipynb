```python
from torchvision import models
import torch
import numpy as np
from torchvision import transforms
```

```python
from torchvision.models.alexnet import Weights
resnet = models.resnet101(pretrained=True)
```

```python
from PIL import Image
data = Image.open("/content/sample_data/cat.jpg")
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
data2 = preprocess(data)
```

```python
tensor = data2.unsqueeze(0)
tensor.shape
```

```python
resnet.eval()
out = resnet(tensor)
out2 = torch.softmax(out, dim=1)
_ , indices = torch.sort(out2, descending=True)
percentages = out2[0] * 100
```
